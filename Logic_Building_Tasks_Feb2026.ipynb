```python
# Pass / Fail Analyzer
marks = [45, 78, 90, 33, 60]
pass_count = 0
fail_count = 0
for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1
        print("Total Passed Students:", pass_count)
print("Total Failed Students:", fail_count)
```

```python
# Simple Data Cleane
names = [" Alice ", "bob", " CHARLIE "]
cleaned_names = [name.strip().lower() for name in names]
print("Cleaned Names:", cleaned_names)
```

```python
#Message Length Analyzer
messages = ["Hi", "Welcome to the platform", "OK"]
for msg in messages:
    length = len(msg)
    print(f"Message: '{msg}' | Length: {length}", end="")
    if length > 10:
        print(" | Flag: Long Message")
    else:
        print(" | Flag: OK")
```
