## Task 2: Multi-Input Prompt System

```python
multi_prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)
test_cases = [
    {
        "topic": "AI",
        "audience": "beginners",
        "tone": "friendly"
    },
    {
        "topic": "Python",
        "audience": "kids",
        "tone": "fun"
    },
    {
        "topic": "Deep Learning",
        "audience": "engineers",
        "tone": "technical"
    }
]
for case in test_cases:
    result = multi_prompt.format(
        topic=case["topic"],
        audience=case["audience"],
        tone=case["tone"]
    )
    print(result)
```

## Task 4: ChatPromptTemplate System

```python
chat_template = ChatPromptTemplate.from_messages([
    (
        "system",
        "You are a helpful {role}."
    ),
    (
        "human",
        "Explain {topic}"
    )
])
# Example Inputs
messages = chat_template.format_messages(
    role="teacher",
    topic="Neural Networks"
)
for message in messages:
    print(message.content)
```

```python
roles = ["teacher", "interviewer", "motivator"]
for role_name in roles:
    print("\nRole:", role_name)
    msgs = chat_template.format_messages(
        role=role_name,
        topic="Artificial Intelligence"
    )
    for msg in msgs:
        print(msg.content)
```

## Task 5: Input Validation Layer

```python
def validate_inputs(audience, tone):
    valid_audience = [
        "beginner",
        "intermediate",
        "expert"
    ]
    valid_tone = [
        "formal",
        "casual",
        "fun"
    ]
    if audience not in valid_audience:
        raise ValueError("Invalid audience value")
    if tone not in valid_tone:
        raise ValueError("Invalid tone value")
    return True
# Test Validation
try:
    validate_inputs("beginner", "fun")
    print("Inputs are valid")
except ValueError as error:
    print(error)
```

## Task 6: Prompt Generator App

```python
def generate_prompt(topic, audience, tone, style):
    templates = {
        "teaching": "Explain {topic} for {audience} in a {tone} teaching style",
        "interview": "Ask interview questions about {topic} for {audience} in a {tone} style",
        "storytelling": "Explain {topic} for {audience} in a {tone} storytelling style"
    }
    prompt_template = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=templates[style]
    )
    final_prompt = prompt_template.format(
        topic=topic,
        audience=audience,
        tone=tone
    )
    return final_prompt
# Example
result = generate_prompt(
    topic="Neural Networks",
    audience="beginners",
    tone="fun",
    style="storytelling"
)
print(result)
```
